```python
import os
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
def getPathName (directory, row):
    fullDirectory = directory + "/" + row['imageName']
    return fullDirectory
```

```python
def createTwoCSV(directory, file):
    df = pd.read_csv(os.path.join(directory, file))
    df['path'] = df.apply(lambda row: getPathName(directory, row), axis=1)
    all_nodules = df[-df['X-min'].isnull()]
    larger_nodules = all_nodules[-(all_nodules['X-min'] == all_nodules['X-max'])]
    return all_nodules, larger_nodules
```

```python
count = 0
final_all_nodules = pd.DataFrame()
final_larger_nodules = pd.DataFrame()
for root, dirs, files in os.walk("/scratch/ebc308/tcia/data/LIDC-IDRI/", topdown=True):
    for name in dirs:
        directory = os.path.join(root, name)
        imagesTogether = []
        xmlTemp = "NA"
        PatientID = directory.split("/")[6]
        for file in os.listdir(directory):
            if file.endswith(".csv"):
                all_nodules, larger_nodules = createTwoCSV(directory, file)
                final_all_nodules = final_all_nodules.append(all_nodules)
                final_larger_nodules = final_larger_nodules.append(larger_nodules)
                count += 1
                print("count: ", count, "Patient ID: ", PatientID)

# final_all_nodules.to_csv("all_nodules.csv")
# final_larger_nodules.to_csv("larger_nodules.csv")
```

```python
cols = list(final_all_nodules.columns)
cols = [cols[-1]] + cols[:-1]
final_all_nodules = final_all_nodules[cols]
final_all_nodules = final_all_nodules.drop(columns='Unnamed: 0')

# getting areas for coords / combine xmin.ymin.xmax.ymax
boxes = []
areas = []
patientids = []
for index, row in final_all_nodules.iterrows():
    # boxes (FloatTensor[N, 4]): the coordinates of the N bounding boxes in [x0, y0, x1, y1] format, 
    # ranging from 0 to W and 0 to H
    # must be list in a list if not compiling boxes
    temp = []
    box = list([row['X-min'],row['Y-min'],row['X-max'],row['Y-max']]) 
    temp.append(box)
    boxes.append(temp)
    area = ((box[2] - box[0]) * (box[3] - box[1]))
    areas.append(area)
    pID = str(row.path).split("/")[6]
    patientids.append(pID)
    
    
final_all_nodules['boxes'] = boxes
final_all_nodules['areas'] = areas
final_all_nodules['patientIDs'] = patientids
final_all_nodules = final_all_nodules.drop(columns=["X-min", "X-max", "Y-min", "Y-max"])
final_all_nodules.head()

final_all_nodules.to_csv("all_nodules.csv", index=False)
```
