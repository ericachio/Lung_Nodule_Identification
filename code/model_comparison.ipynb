```python
import pickle
import matplotlib.pyplot as plt
```

## Faster RCNN

```python
faster_100area = pickle.load(open("pickle/fasterrcnn_resnet50_area100_coco_validate_acc.pkl", "rb" ) )
faster_50area = pickle.load(open("pickle/fasterrcnn_resnet50_area50_coco_validate_acc.pkl", "rb"))
faster_ImageNet = pickle.load(open("pickle/fasterrcnn_resnet50_area100_imageNet_validate_acc.pkl", "rb"))
```

```python
plt.title('Validation Accuracies for Faster-RCNN pretrained on COCO vs ImageNet')
plt.plot(faster_100area, label = "COCO")
plt.plot(faster_ImageNet, label = "ImageNet")
plt.legend(loc = "lower right")
```

```python
plt.title('Validation Accuracies for Faster-RCNN on Nodule Area >=50 vs >=100')
plt.plot(faster_100area, label = "Area >=100")
plt.plot(faster_50area, label = "Area >=50")
plt.legend(loc = "lower right")
```
